```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np


def plot_avgs(df, axes, format=None):
  df_avgs = df
  # take averages
  df_avgs['High'] = np.log(df_avgs['High'].rolling(14).mean())
  # df_avgs['Low'] = np.log(df_avgs['Low'].rolling(30).mean())
  # df_avgs['Close'] = np.log(df_avgs['Close'].rolling(30).mean())
  # df_avgs['Open'] = np.log(df_avgs['Open'].rolling(30).mean())
  figure = df_avgs.plot(x='Date', y=['High'], kind='line', figsize=(15,10), ax=axes, style=format)
  figure.tick_params(labelrotation=45)
  return 0


def read_data(filename):
  df = pd.read_csv(filename, thousands=',')
  # clean the data
  df = df.dropna(how='any')
  return df


def correct_dtypes(df):
  Df = df
  Df['Date'] = pd.to_datetime(Df['Date'])
  Df['High'] = Df['High'].astype(float)
  # Df['Low'] = Df['Low'].astype(float)
  # Df['Open'] = Df['Open'].astype(float)
  # Df['Close'] = Df['Close'].astype(float)
  return Df


# if time remains account for the difference in currencies, since rates would be
# different in usd and inr
df_bitcoin = correct_dtypes(read_data('coin_Bitcoin.csv'))
df_ethereum = correct_dtypes(read_data('coin_Ethereum.csv'))
df_dogecoin = correct_dtypes(read_data('coin_Dogecoin.csv'))
df_polka = correct_dtypes(read_data('coin_Polkadot.csv'))
df_lite = correct_dtypes(read_data('coin_Litecoin.csv'))
df_nifty50 = correct_dtypes(read_data('Nifty50.csv'))
df_snp500 = correct_dtypes(read_data('S&P500.csv'))
df_snp500.rename(columns={'Close/Last':'Close'}, inplace=True)
df_dowjones = correct_dtypes(read_data('DowJones.csv'))
df_snp500.rename(columns={'Price':'Close'}, inplace=True)

# plotting averages because I want to look at smoother graphs
ax = plt.axes()
plot_avgs(df_bitcoin, ax)
plot_avgs(df_ethereum, ax)
plot_avgs(df_dogecoin, ax)
plot_avgs(df_polka, ax)
plot_avgs(df_lite, ax)
plot_avgs(df_nifty50, ax, 'c--')
plot_avgs(df_snp500, ax, 'm--')
plot_avgs(df_dowjones, ax, 'y--')
ax.legend(['bitcoin', 'ethereum', 'dogecoin', 'polkadot', 'litecoin', 'Nifty-50', 'S&P500', 'Dow Jones'])
ax.set_title('Market Price (log scale)')
ax.set_ylabel('Price')
```

Training the data set by 80-20 rule \\
80% - Training data \\
20% - Test data

We will try this on only the 80% of total data from beginning and predict the closing prices of the last remaining 20% of the data.

The features of this training may be the volume traded, open price, low and high values, and market cap

```python
#Using the prices from the Close column in our dataset
close_prices = df.filter(['Close'])
close_prices

#Organizing the prices into an array
df_close_price = close_prices.values

#Setting the length the training dataset into 80% of the total data
#Ceiling used to round up to the nearest whole int
train_len = math.ceil(len(df_close_price) * 0.80)

scaler = MinMaxScaler(feature_range=(0,1))

#reshaping the data into two-dimensions (a list of lists of values)
scaled_data = scaler.fit_transform(df_close_price)

#Training data
train_data = scaled_data[0: train_len, :]
train_data



```

Training

```python
train_x = [] #features
train_y = [] #target

len_train = len(train_data)

for i in range(60, len_train):
    train_data_list_x = train_data[i-60:i, 0] #each time gives list of previous 60 prices to predict the 61th price from beginning to end
    train_x.append(train_data_list_x)

    train_data_list_y = train_data[i, 0] #61th price value till end
    train_y.append(train_data_list_y)

#Convert lists -> arrays since we will be using it to train models

train_x, train_y = np.array(train_x), np.array(train_y)

#Converting them to 3-dimensions which will be used for model

train_x = np.reshape(train_x, (train_x.shape[0], train_x.shape[1],1)) #here 1 refers to the number of features we are referring to i.e the closing price

```

Building LSTM Model

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM #Long short-term memory

model = Sequential() #The model is named Sequential


#inputs: A 3D tensor with shape `[batch, timesteps, feature]`.
model.add(LSTM(40, return_sequences=True, input_shape=(train_x.shape[1],1)))
model.add(LSTM(30, return_sequences=False))
model.add(Dense(85))
model.add(Dense(1)) #We want one output
model.summary()
```

Model compiling

```python
#Adam optimization is a stochastic gradient descent method,
#Loss: RMS Error
model.compile(optimizer='adam', loss= 'mean_squared_error')

model.fit(train_x,train_y,batch_size=1, epochs=1)
```

Making test data

```python
from keras.engine.sequential import track_nodes_created_by_last_call
data_for_test = scaled_data[train_len-60:,:]

test_x = []

test_y = df_close_price[train_len:,:]

for i in range(60,len(data_for_test)):
    test_x.append(data_for_test[i-60:i,0])

#Convert test values into array then reshape for it to be input into model
test_x = np.array(test_x)
test_x = np.reshape(test_x, (test_x.shape[0],test_x.shape[1],1))
```

Testing model and predicting values

```python
predict_y = model.predict(test_x)

predict_y = scaler.inverse_transform(predict_y)
```

Root Mean Square Error (RMSE)

```python
rmse = np.sqrt(np.mean(((predict_y - test_y)**2)))
print(rmse)
```

Training the data set by 80-20 rule \\
80% - Training data \\
20% - Test data

We will try this on only the 80% of total data from beginning and predict the closing prices of the last remaining 20% of the data.

The features of this training may be the volume traded, open price, low and high values, and market cap

```python
#Using the prices from the Close column in our dataset
close_prices = df.filter(['Close'])
close_prices

#Organizing the prices into an array
df_close_price = close_prices.values

#Setting the length the training dataset into 80% of the total data
#Ceiling used to round up to the nearest whole int
train_len = math.ceil(len(df_close_price) * 0.80)

scaler = MinMaxScaler(feature_range=(0,1))

#reshaping the data into two-dimensions (a list of lists of values)
scaled_data = scaler.fit_transform(df_close_price)

#Training data
train_data = scaled_data[0: train_len, :]
train_data



```

Training

```python
train_x = [] #features
train_y = [] #target

len_train = len(train_data)

for i in range(60, len_train):
    train_data_list_x = train_data[i-60:i, 0] #each time gives list of previous 60 prices to predict the 61th price from beginning to end
    train_x.append(train_data_list_x)

    train_data_list_y = train_data[i, 0] #61th price value till end
    train_y.append(train_data_list_y)

#Convert lists -> arrays since we will be using it to train models

train_x, train_y = np.array(train_x), np.array(train_y)

#Converting them to 3-dimensions which will be used for model

train_x = np.reshape(train_x, (train_x.shape[0], train_x.shape[1],1)) #here 1 refers to the number of features we are referring to i.e the closing price

```

Building LSTM Model

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM #Long short-term memory

model = Sequential() #The model is named Sequential


#inputs: A 3D tensor with shape `[batch, timesteps, feature]`.
model.add(LSTM(40, return_sequences=True, input_shape=(train_x.shape[1],1)))
model.add(LSTM(30, return_sequences=False))
model.add(Dense(85))
model.add(Dense(1)) #We want one output
model.summary()
```

Model compiling

```python
#Adam optimization is a stochastic gradient descent method,
#Loss: RMS Error
model.compile(optimizer='adam', loss= 'mean_squared_error')

model.fit(train_x,train_y,batch_size=1, epochs=1)
```

Making test data

```python
from keras.engine.sequential import track_nodes_created_by_last_call
data_for_test = scaled_data[train_len-60:,:]

test_x = []

test_y = df_close_price[train_len:,:]

for i in range(60,len(data_for_test)):
    test_x.append(data_for_test[i-60:i,0])

#Convert test values into array then reshape for it to be input into model
test_x = np.array(test_x)
test_x = np.reshape(test_x, (test_x.shape[0],test_x.shape[1],1))
```

Testing model and predicting values

```python
predict_y = model.predict(test_x)

predict_y = scaler.inverse_transform(predict_y)
```

Root Mean Square Error (RMSE)

```python
rmse = np.sqrt(np.mean(((predict_y - test_y)**2)))
print(rmse)
```
